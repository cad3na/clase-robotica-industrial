```python
def matrizDH(th1, a1, d1, al1):
    return Matrix([[cos(th1), -1*sin(th1)*cos(al1), sin(th1)*sin(al1), a1*cos(th1)],
                   [sin(th1), cos(th1)*cos(al1), -1*cos(th1)*sin(al1), a1*sin(th1)],
                   [0, sin(al1), cos(al1), d1],
                   [0, 0, 0, 1]])
```
